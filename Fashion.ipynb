```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import keras
(train_input, train_target), (test_input, test_target) =\
keras.datasets.fashion_mnist.load_data()
```

## STEP #3: TRAINING THE MODEL

```python
#  
train_input = train_input / 255
test_input = test_input / 255
```

```python
from sklearn.model_selection import train_test_split
```

```python
#         
train_input, val_input, train_target, val_target = train_test_split(train_input, train_target, test_size=0.2, random_state=12345)
```

```python
train_input = train_input.reshape(train_input.shape[0], *(28,28,1))
```

```python
#  1     .
train_input.shape
```

```python
val_input = val_input.reshape(val_input.shape[0], *(28,28,1))
test_input = test_input.reshape(test_input.shape[0], *(28,28,1))
```

```python
val_input.shape
```

```python
test_input.shape
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam
from keras.callbacks import TensorBoard
```

```python
cnn_model = Sequential()
```

```python
#   32, 3X3,   28X28   ,   ReLU 
cnn_model.add(Conv2D(64,3,3, input_shape = (28,28,1), activation='relu'))
```

```python
cnn_model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
cnn_model.add(Flatten())
```

```python
cnn_model.add(Dense(32, activation = 'relu'))
```

```python
cnn_model.add(Dense(10, activation = 'sigmoid'))
```

```python
cnn_model.compile(loss='sparse_categorical_crossentropy', optimizer=Adam(learning_rate=0.001), metrics=['accuracy'])
```

```python
cnn_model.fit(train_input, train_target, batch_size = 512, epochs = 50, verbose = 1, validation_data = (val_input, val_target))
```

## STEP #5: EVALUATING THE MODEL

```python
evaluation = cnn_model.evaluate(test_input, test_target)
print('Test Accuracy : {}'.format(evaluation[1]))
```

```python
predicted_classes = cnn_model.predict(test_input)
```

```python
predicted_classes = np.argmax(predicted_classes, axis = 1)
```

```python
L = 5
W = 5

fig,axes = plt.subplots(L, W, figsize = (12,12))

axes = axes.ravel()

for i in np.arange(0, L*W):
  axes[i].imshow(test_input[i].reshape(28,28))
  axes[i].set_title("Prediction Class = {}\n True Class = {}".format(predicted_classes[i], test_target[i]))
  axes[i].axis('off')

plt.subplots_adjust(hspace=0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(test_target, predicted_classes)
plt.figure(figsize = (14,10))
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import classification_report

num_classes = 10
target_names = ["Class {}".format(i) for i in range(num_classes)]

print(classification_report(test_target, predicted_classes, target_names=target_names))
```
